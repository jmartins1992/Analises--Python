# grafico 3d - grass

```python
import pandas as pd

arquivo1 = pd.ExcelFile('resultados1.xlsx')
dadosG1 = pd.read_excel(arquivo1,'GRAMA3D', sep=',')
dadosM1 = pd.read_excel(arquivo1,'mulch3d', sep=',')
dadosB1 = pd.read_excel(arquivo1,'bared3d', sep=',')

dadosB1['y'].describe()
```

```python
import statsmodels.api as sm

#acertando as variáveis
yobs= dadosG1['x1']
xobs=dadosG1[['y1','z1']]
xobs=sm.add_constant(xobs)
resultadoG=sm.OLS(yobs,xobs).fit()
print(resultadoG.summary())
```

```python

from mpl_toolkits.mplot3d import Axes3D

import matplotlib.pyplot as plt

import seaborn as sns


fig = plt.figure(figsize=(10,10))
ax= fig.gca(projection='3d')
surf = ax.plot_trisurf(dadosG1['y1'],dadosG1['x1'],dadosG1['z1'],cmap=plt.cm.coolwarm)

fig.colorbar(surf,shrink=0.5,aspect=10)

ax.view_init(60,125)

ax.set_xlabel('Temperature Surface °C')
ax.set_ylabel('Soil Temperature °C')
ax.set_zlabel('Depth (cm)')

clin = resultadoG.params[0]
cang1 = resultadoG.params[1]
cang2 = resultadoG.params[2]
r2aj = resultadoG.rsquared_adj
pvalor = resultadoG.f_pvalue

texto = " XObs = {0:.3f} .{1:.3f} .{2:.3f} \n R2 = {3:.2f} \n P-Valor = {4:.3E}".format(cang1,cang2,clin,r2aj,pvalor)

ax.text(10,10,45,texto,fontsize=10)
```
