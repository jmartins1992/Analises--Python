```python
import pandas as pd

dadosTeste = pd.read_csv("carbono.csv", sep=",")

dadosTeste.transpose()#Comando para exibir o DataFrame transposto
```

```python
dadosTeste.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

sns.set(style="whitegrid", color_codes=True)

iris = dadosTeste

 
sns.boxplot(data=iris);
```

```python
sns.set(style="whitegrid")

# atrubuindo os dataframe a x e y
x = dadosTeste['SOC']
y = dadosTeste['Sand']

# função responsavel por contruir o grafico
sns.residplot(x, y, lowess=True, color="b")
```

```python
sns.set(style="whitegrid")

# atrubuindo os dataframe a x e y

x = dadosTeste['SOC']
y = dadosTeste['Silt']

# função responsavel por contruir o grafico
sns.residplot(x, y, lowess=True, color="g")
```

```python

sns.set(style="ticks")

x = dadosTeste['SOC']
y = dadosTeste['Sand']

sns.jointplot(x, y, kind="reg", color="b")
```

```python
sns.set(style="ticks")

x = dadosTeste['SOC']
y = dadosTeste['Silt']

# função responsavel por contruir o grafico
sns.jointplot(x, y, kind="reg", color="g")
```

```python
from string import ascii_letters
import numpy as np

sns.set(style="white")

# Passando para a variavel d o data frame dos dados de carbono organico

d = dadosTeste

# gerando a correlação da matrix atravez da função .corr()
corr = d.corr()

#cria as mascaras para o topo do triangulo
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

f, ax = plt.subplots(figsize=(11, 9))

# função para gerar a cor do dados do gráfico
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# função que recebe a correlação e a mascara superior e todos os outros atributos e constroi o grafico
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
