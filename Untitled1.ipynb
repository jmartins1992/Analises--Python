```python
import pandas as pd
wine = pd.read_csv('wine_data.csv',names=["Cultivator","Alchol","Malic_Acid","Alcalinity_of_Ash",
                                        "Magnesium","Total_phenols","Falvanoids","Noflavanoid_phenols",
                                        "Proanthocyanins","Color_intensity","Hue","OD280","Proline"])

wine.describe().transpose()
```

```python
#variavel depenente y

X = wine.drop('Cultivator',axis=1)#dados referentes a todo objeto menos a primeira linha
y= wine['Cultivator']#dados referentes a primeira coluna
```

```python
# separando os dados para o teste 

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
#padronizar os dados

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```
