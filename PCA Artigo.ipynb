```python
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.cluster import KMeans

dados.columns = map(str.upper, dados.columns)

dados_clean = dados.dropna()

#print(dados_clean)

cluster = dados_clean[['FLIR','X2','X5','X10','X20','X30','TMAX','TMIN','TMED']]

clustervar = cluster.copy()

clustervar['FLIR'] = preprocessing.scale(clustervar['FLIR'].astype('float64'))

clustervar['FLIR'] = preprocessing.scale(clustervar['X2'].astype('float64'))

clustervar['FLIR'] = preprocessing.scale(clustervar['X5'].astype('float64'))

clustervar['FLIR'] = preprocessing.scale(clustervar['X10'].astype('float64'))

clustervar['FLIR'] = preprocessing.scale(clustervar['X20'].astype('float64'))

clustervar['FLIR'] = preprocessing.scale(clustervar['X30'].astype('float64'))

clustervar['FLIR'] = preprocessing.scale(clustervar['TMAX'].astype('float64'))

clustervar['FLIR'] = preprocessing.scale(clustervar['TMIN'].astype('float64'))

clustervar['FLIR'] = preprocessing.scale(clustervar['TMED'].astype('float64'))

clus_train, clus_test = train_test_split(clustervar, test_size=.3, random_state=123)

from scipy.spatial.distance import cdist

cluster = range(1,100)

meandist = []

for k in cluster:
    model = KMeans(n_clusters=k)
    model.fit(clus_train)
    clusassign = model.predict(clus_train)
    meandist.append(sum(np.min(cdist(clus_train, model.cluster_centers_, 'euclidean'),axis=1))/clus_train.shape[0])
    

plt.plot(cluster,meandist)
plt.xlabel('Numero de clusters')
plt.ylabel('Média da Distancia')
```

```python
mode13 = KMeans(n_clusters=9)
mode13.fit(clus_train)
clausassign = mode13.predict(clus_train)

from sklearn.decomposition import PCA

pca_2 = PCA(9)

plot_columns = pca_2.fit_transform(clus_train)

plt.scatter(x=plot_columns[:,0],y=plot_columns[:,1], c=mode13.labels_,)

plt.xlabel('Variavel canonica 1')
plt.ylabel('Variavel canonica 2')
plt.show()
```
